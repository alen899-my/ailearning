```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
data=pd.read_csv("height-weight.csv")
data
```

```python
#scatter plot
plt.scatter(data["Weight"],data["Height"])
plt.xlabel("Weight")
plt.ylabel("Height")
```

```python
##correlation
data.corr()
```

```python
#seaborn
import seaborn as sns
sns.pairplot(data)
```

```python
##divide into independed and depenedede
data.head()
```

```python
X=data[["Weight"]]#indidpended feature should be dataframe or 2 dimentionalarrau
X.head()
```

```python
y=data["Height"]#series is better
y.head()
```

```python
###train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
#apply linear reggression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression(n_jobs=-1)
```

```python
regression.fit(X_train,y_train)
```

```python
coef=regression.coef_
interspt=regression.intercept_
```

```python
print("Coefficient or slope",coef)
print("intersept",interspt)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,regression.predict(X_train))
```

```python
##prediction test data
```

```python
y_prediction=regression.predict(X_test)
y_prediction
```

prediction of test data
1.predicted height output=intercept+coef(weight)
2.y_predict_test=156.47+17.29(x_test)

```python
#perforamance
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_prediction)
mae=mean_absolute_error(y_test,y_prediction)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```
