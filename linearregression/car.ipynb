```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("cardata.csv", encoding="cp1252")
data
```

```python
data["Company Names"].unique()
data["Company Names"]=data["Company Names"].str.upper().str.strip()
data["Company Names"].unique()
```

```python
from sklearn.preprocessing import  OneHotEncoder,LabelEncoder
```

```python
encoder=OneHotEncoder()
```

```python
encoded=encoder.fit_transform(data[["Company Names"]]).toarray()
encoded
dataframe=pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
dataframe.info()
```

```python
data["Cars Names"].unique()
```

```python
lbl_encoder=LabelEncoder()
car_name_encoded=lbl_encoder.fit_transform(data["Cars Names"])
data["Car Names Encoded"]=car_name_encoded
data[["Cars Names","Car Names Encoded"]].head(10)
```

```python
updated_data=pd.concat([data,dataframe],axis=1)
updated_data.drop(columns=["Company Names","Cars Names"],axis=1,inplace=True)
updated_data
```

```python
import re
def extract_engine_size(engine):
    engine = str(engine).upper()
    # split by commas and spaces
    parts = re.split("[ ,]", engine)  
    for part in parts:
        if "L" in part:
            try:
                # remove L and convert to float
                return float(part.replace("L",""))
            except:
                pass
    return None

# apply to your column
updated_data["Engine_Size_L"] = updated_data["Engines"].apply(extract_engine_size)

# check results
updated_data[["Engines", "Engine_Size_L"]].head(10)


        
```
