```python
import re
def clean_hp(value):
    if pd.isna(value):
        return None
    numbers=re.findall(r"\d+",str(value))
    if not numbers:
        return None
    if len(numbers)==2:
        return (int(numbers[0])+int(numbers[1]))/2
    return int(numbers[0])
updated_data["Hp_encoded"]=updated_data["HorsePower"].apply(clean_hp)
updated_data["Hp_encoded"]
```

```python
def format_performance(value):
    if pd.isna(value):
        return None
    performance=re.findall(r"\d+\.?\d*",str(value))
    if not performance:
        return None
    if len(performance)==2:
        return (float(performance[0])+float(performance[1]))/2
    return float(performance[0])
updated_data["performance_encoded"]=updated_data["Performance(0 - 100 )KM/H"].apply(format_performance)
```

```python
def clean_price(value):
    if pd.isna(value):
        return None
    text=str(value)
    text=text.replace("â‚¬","").replace("$","")
    text = text.replace("Â–", "-")       # fix special dash
    text = text.strip()       
    numbers = re.findall(r"[0-9,]+", str(text))
    numbers=[float(num.replace(",",""))for num in numbers]
    if not numbers:
        return None
    if len(numbers)==2:
        return sum(numbers)/2
    return numbers[0]
```

```python
updated_data["cleaned_Car_prices"]=updated_data["Cars Prices"].apply(clean_price)
```

```python
def clean_fuels(values):
    text=str(values).lower().strip()

    text=text.replace(",","/").replace("+","/").replace("(","").replace(")","").replace("/","").replace("-","")
    text=text.replace(" ","")
    text=text.replace("hyrbrid","hybrid")
     # standardize common terms
    text = text.replace("plug-inhybrid", "plugin-hybrid")
    text = text.replace("hybrid/petrol", "petrol-hybrid")
    text = text.replace("petrolhybrid", "petrol-hybrid")
    text = text.replace("dieselhybrid", "diesel-hybrid")
    text = text.replace("hybrid/gaselectric", "hybrid-electric")
    text=text.replace("plugin-hybrid","pluginhybrid")
    text=text.replace("hybrid/plug-in","hybridplug-in")  
    text=text.replace("gas/hybrid","gashybrid")  
    text=text.replace("petrol-hybrid","petrolhybrid")  
    text=text.replace("cng/petrol","cngpetrol")  
    text=text.replace("hybridgas/electric","hybridgaselectric")  
    text=text.replace("diesel-hybrid","dieselhybrid")
    return text
```

```python
updated_data["Clean_fuels"]=updated_data["Fuel Types"].apply(clean_fuels)
```

```python
def clean_engines(values):
    text=str(values).lower().strip()

    #extracting displacement
    displacement_match = re.search(r"(\d+(?:\.\d+)?)\s*l", text)

    if displacement_match:
        displacement=float(displacement_match.group(1))
    else:
        displacement=np.nan

    config_match=re.search(r"(v\d|i\d|boxer-\d|w\d+|wankel|rotary|inline-\d)",text)
    if config_match:
        config=config_match.group(0).upper()
    else:
        config=None
    if "electric" in text:
        powertrain = "Electric"
    elif "hybrid" in text or "plug" in text:
        powertrain = "Hybrid"
    elif "diesel" in text:
        powertrain = "Diesel"
    elif "petrol" in text or "gas" in text:
        powertrain = "Petrol"
    elif "hydrogen" in text:
        powertrain = "Hydrogen"
    else:
        powertrain = "Other"
    return pd.Series([displacement,config,powertrain])

  
updated_data[["Engine_displacement","Engine_config","PowerTrain"]]=updated_data["Engines"].apply(clean_engines)
```
